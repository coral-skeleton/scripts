```python
    import matplotlib.pyplot as plt
    %matplotlib inline
    import numpy as np
```

```python
masses = np.array([10.11,10.28,8.86,9.03,8.98,9.02,9.02,7.89,8.28])
names = np.array(['Mkt11','Mkt12','Mkt13','Mkt14','Mkt15','Mkt16','Mkt17','Mkt18','Mkt19'])
errmass = np.array([0.1,0.1,0.12,0.12,0.11,0.1,0.11,0.23,0.17])
y_pos = np.arange(len(masses))
plt.figure(figsize = (8,5))
bars = plt.bar(y_pos, masses, yerr= errmass, capsize = 19, label = 'results', zorder = 2)
#plt.bar(y_pos+width/2, dzudmass,width, yerr= errdzud, capsize = 5, color = '#00fc97', label = 'Dzudzar')
plt.xticks(y_pos, names, fontsize = 14)
plt.yticks(fontsize = 16)
#plt.legend(fontsize = 16)
plt.ylabel(r'logM$_{HI}$ [M$_\odot$]', fontsize = 18)
plt.title('HI Masses of new galaxies', fontsize = 20)
plt.ylim(7,10.5)
#plt.savefig('newmasses.png',bbox_inches = 'tight')
```

```python
R = np.linspace(1,50, num = 200)
V = -1/R + 1
V2 = V+0.1
V3 = V-0.1

fig = plt.figure(figsize = (12,5))
plt.plot(R,V, label = 'Disk Component',lw = 3)
plt.plot(R,V2, label = 'Leading Component', lw = 3)
plt.plot(R,V3, label = 'Lagging Component', lw = 3)
plt.legend(fontsize = 18)
plt.ylabel('Rotational Velocity [km/s]', fontsize = 18)
plt.xlabel('Radius [kpc]', fontsize = 18)

plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False)

plt.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=False,      # ticks along the bottom edge are off
    right=False,         # ticks along the top edge are off
    labelleft=False)
plt.ylim(0.5,1.5)
plt.xlim(1,55)


#plt.savefig('rotation.png', bbox_inches='tight')
```
